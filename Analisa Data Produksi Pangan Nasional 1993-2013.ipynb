# 2. Statistik Data Produksi Pangan Nasional

```python
#menghitung  jumlah, rata-rata, maksimum serta minimum data luas panen
total1 = data_pangan.luas_panen.sum()
mean1 = data_pangan.luas_panen.mean()
max1 = data_pangan.luas_panen.max()
min1 = data_pangan.luas_panen.min()

print("Total luasan panen padi periode 1993-2013 adalah: ", total1, "Hektar")
print("Rata-rata luasan panen padi provinsi periode 1993-2013 adalah: ", mean1, "Hektar")
print("Maksimum luasan panen padi provinsi periode 1993-2013 adalah: ", max1, "Hektar")
print("Minimum luasan panen padi provinsi periode 1993-2013 adalah: ", min1, "Hektar")
```

```python
#menghitung  jumlah, rata-rata, maksimum serta minimum data produksi
total2 = data_pangan.produksi.sum()
mean2 = data_pangan.produksi.mean()
max2 = data_pangan.produksi.max()
min2 = data_pangan.produksi.min()

print("Total produksi padi periode 1993-2013 adalah: ", total2, "Ton")
print("Rata-rata produksi padi provinsi periode 1993-2013 adalah: ", mean2, "Ton")
print("Maksimum produksi padi provinsi periode 1993-2013 adalah: ", max2, "Ton")
print("Minimum produksi padi provinsi periode 1993-2013 adalah: ", min2, "Ton")
```

```python
#menghitung  rata-rata, maksimum serta minimum data luas panen
mean3 = data_pangan.produktivitas.mean()
max3 = data_pangan.produktivitas.max()
min3 = data_pangan.produktivitas.min()


print("Rata-rata produktivitas padi provinsi periode 1993-2013 adalah: ", mean3, "Ton/Hektar")
print("Maksimum produktivitas padi provinsi periode 1993-2013 adalah: ", max3, "Ton/Hektar")
print("Minimum produktivitas padi provinsi periode 1993-2013 adalah: ", min3, "Ton/Hektar")
```

# 3. Visualisasi

```python
#luas_panen nasional
ax = data_pangan.groupby('tahun').mean()['luas_panen']
ax = ax.plot(kind='line', title ="Grafik 1: Rata-rata Luas Panen Padi Nasional 1993-2013",figsize=(15,5), color='green', linewidth=4, legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Luas Panen (Hektar)", fontsize=12)
```

```python
#produksi padi nasional
ax = data_pangan.groupby('tahun').mean()['produksi']
ax = ax.plot(kind='line', title ="Grafik 2: Rata-rata Produksi Padi Nasional 1993-2013",figsize=(15,5), color='blue', linewidth=4, legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Produksi (Ton)", fontsize=12)
```

```python
#produktivitas padi nasional
ax = data_pangan.groupby('tahun').mean()['produktivitas']
ax = ax.plot(kind='line', title ="Grafik 3: Rata-rata Produktivitas Padi Nasional 1993-2013",figsize=(15,5), color='olive', linewidth=4, legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Produktivitas (Ton/Hektar)", fontsize=12)
```

```python
#luas_panen per provinsi
ax = data_pangan.groupby('nama_provinsi').mean()['luas_panen']
ax = ax.plot(kind='bar', title ="Grafik 4: Rata-rata Luas Panen Padi per Provinsi 1993-2013",figsize=(15,5), linewidth=4, legend=True, fontsize=12)
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Luas Panen (Hektar)", fontsize=12)
```

```python
#produksi per provinsi
ax = data_pangan.groupby('nama_provinsi').mean()['produksi']
ax = ax.plot(kind='bar', title ="Grafik 5: Rata-rata Produksi Padi per Provinsi 1993-2013",figsize=(15,5), linewidth=4, legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Produksi (Ton)", fontsize=12)
```

```python
#produktivitas per Provinsi
ax = data_pangan.groupby('nama_provinsi').mean()['produktivitas']
ax = ax.plot(kind='bar', title ="Grafik 3: Rata-rata Produktivitas per Provinsi 1993-2013",figsize=(12,5), legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Produktivitas (Ton/Hektar)", fontsize=12)
```
